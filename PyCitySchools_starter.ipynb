# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of schools
schnum=len(school_data_complete["school_name"].unique())
# Calculate the total number of students
stunum=len(school_data_complete["student_name"])
# Calculate the average math score
mathsc=sum(school_data_complete["math_score"])/stunum
# Calculate the average reading score
readsc=sum(school_data_complete["reading_score"])/stunum
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
ovrsc=(mathsc+readsc)/2
#Calculate the percentage of students with a passing math score (70 or greater)
pass_math = len(school_data_complete.loc[school_data_complete["math_score"] >= 70, :])*100/stunum
#Calculate the percentage of students with a passing reading score (70 or greater)
pass_read=len(school_data_complete.loc[school_data_complete["reading_score"] >= 70, :])*100/stunum
#Calculate the total budget
school_budg=sum(school_data_complete["budget"].unique())
#Create a dataframe to hold the above results
df_summary=pd.DataFrame([{"Total Schools": schnum
                          , "Total Students": stunum
                          , "Total Budget": school_budg
                          , "Average Math Score": mathsc
                          , "Average Reading Score": readsc
                          , "% Passing Math": pass_math
                          , "% Passing Reading": pass_read
                          , "% Overall Passing Rate": ovrsc}])
#displayed data cleaner formatting
df_summary["% Passing Math"]=df_summary["% Passing Math"].map("{:.2f}%".format)
df_summary["% Passing Reading"]=df_summary["% Passing Reading"].map("{:.2f}%".format)
df_summary["% Overall Passing Rate"]=df_summary["% Overall Passing Rate"].map("{:.2f}%".format)
df_summary["Average Math Score"]=df_summary["Average Math Score"].map("{:.2f}".format)
df_summary["Average Reading Score"]=df_summary["Average Reading Score"].map("{:.2f}".format)
df_summary=df_summary[["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
df_summary
```

* Sort and display the top five schools in overall passing rate

```python
#overview table that summarizes key metrics about each school
school_gb=school_data_complete.groupby("school_name")
type=school_gb["type"].max()
#Total Students
totstudents=school_gb["student_name"].count()
#Total School Budget
budget=school_gb["budget"].max()
#Per Student Budget
bdt_per_stu=budget/totstudents
#Average Math Score
avg_math=school_gb["math_score"].mean()
#Average Reading Score
avg_read=school_gb["reading_score"].mean()
#calculation for % passing math and reading
#capture dataframe for each population
perMath=school_data_complete.loc[school_data_complete["math_score"]>=70, :]
perRead=school_data_complete.loc[school_data_complete["reading_score"]>=70, :]
#group by for each school and population of each school
school_m_gb= perMath.groupby("school_name")
school_r_gb=perRead.groupby("school_name")
passMath=school_m_gb["student_name"].count()
passRead=school_r_gb["student_name"].count()
school_summary_df=pd.DataFrame({"School Type": type
                                , "Total Students": totstudents
                               , "Total School Budget": budget
                               , "Per Student Budget": budget/totstudents
                               , "Average Math Score": avg_math
                               , "Average Reading Score": avg_read
                                #final calc for percent passing
                               , "% Passing Math": passMath*100/totstudents
                               , "% Passing Reading": passRead*100/totstudents
                                #total percent passing
                               , "% Overall Passing Rate": ((passMath/totstudents)+(passRead/totstudents))/2 * 100})
#formatting
school_summary_df["Average Math Score"]=school_summary_df["Average Math Score"].map("{:.2f}%".format)
school_summary_df["Average Reading Score"]=school_summary_df["Average Reading Score"].map("{:.2f}%".format)
school_summary_df["% Passing Math"]=school_summary_df["% Passing Math"].map("{:.2f}%".format)
school_summary_df["% Passing Reading"]=school_summary_df["% Passing Reading"].map("{:.2f}%".format)
school_summary_df["% Overall Passing Rate"]=school_summary_df["% Overall Passing Rate"].map("{:.2f}%".format)
#sorting
school_sort_best=school_summary_df.sort_values("% Overall Passing Rate",ascending=False)
school_sort_best.head()
```

```python
#Bottom Performing Schools (By Passing Rate)
#Sort and display the five worst-performing schools
school_sort_worst=school_summary_df.sort_values("% Overall Passing Rate")
school_sort_worst.head()
```

```python
#Math Scores by Grade breakdown
# pandas for each grade
nine_df=school_data_complete.loc[school_data_complete["grade"]=="9th", :]
ten_df=school_data_complete.loc[school_data_complete["grade"]=="10th", :]
eleven_df=school_data_complete.loc[school_data_complete["grade"]=="11th", :]
twelve_df=school_data_complete.loc[school_data_complete["grade"]=="12th", :]
nine_gb=nine_df.groupby("school_name")
ten_gb=ten_df.groupby("school_name")
eleven_gb=eleven_df.groupby("school_name")
twelve_gb=twelve_df.groupby("school_name")
nine_score=nine_gb["math_score"].mean()
ten_score=ten_gb["math_score"].mean()
eleven_score=eleven_gb["math_score"].mean()
twelve_score=twelve_gb["math_score"].mean()
#Combine the series into a dataframe
math_df=pd.DataFrame({"9th": nine_score
                     , "10th": ten_score
                     , "11th": eleven_score
                     , "12th": twelve_score})
#formatting
math_df["9th"]=math_df["9th"].map("{:.2f}%".format)
math_df["10th"]=math_df["10th"].map("{:.2f}%".format)
math_df["11th"]=math_df["11th"].map("{:.2f}%".format)
math_df["12th"]=math_df["12th"].map("{:.2f}%".format)
math_df
```

```python
#Reading Score by Grade
#breakdown reading for each grade
nine_rscore=nine_gb["reading_score"].mean()
ten_rscore=ten_gb["reading_score"].mean()
eleven_rscore=eleven_gb["reading_score"].mean()
twelve_rscore=twelve_gb["reading_score"].mean()
# combine into data frame
read_df=pd.DataFrame({"9th": nine_score
                     , "10th": ten_score
                     , "11th": eleven_score
                     , "12th": twelve_score})
#formatting
read_df["9th"]=read_df["9th"].map("{:.2f}%".format)
read_df["10th"]=read_df["10th"].map("{:.2f}%".format)
read_df["11th"]=read_df["11th"].map("{:.2f}%".format)
read_df["12th"]=read_df["12th"].map("{:.2f}%".format)
read_df
```

```python
#Table to breakdown grades per spending based off reasonable allocations
# bins creation
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_summary_df["Spending Ranges (Per Student)"]=pd.cut(school_summary_df["Per Student Budget"], spending_bins, labels=group_names)
#group by based off previous dataframe
school_summary_gb=school_summary_df.groupby("Spending Ranges (Per Student)")

sch_summ_df=school_summary_gb.max()

sch_summ_fin=sch_summ_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
sch_summ_fin
```
